# Part 3: Making My Own Data Visualization Map

Given that we have crime data available to us, a natural data visualization task from a law enforcement perspective would be to see if we have clustering of crimes, ie. do reported murders often occur in the "same part of town". Hence, we will be mapping every crime that has occurred within our data set in the Baltimore region to see if there are any clustering patterns that exist.

## Designating Serious vs Nonserious Crime

First, we have to determine what types of crimes are serious or not. For us here, we will denote **murder**, **theft with weapon**, **assault/abuse**, and **sex-related** crimes to be serious crimes, while others for example **narcotics** related crimes to be more nonserious as a lot oft hose incidents are from possession of *marijuana* which is in the process of being legalized. This choice should be made because crimes that cause bodily harm and/or serious mental distress can have a longer lasting impact on people as compared to consuming marijuana or other minor drugs, and hence for initial differentiation, we should consider a dichotomy between "serious" and "nonserious" crime.

```python
def differentiator(crime_type):
    serious_crime = False
    
    # denoting sex related crime
    if 'Sex' in crime_type or 'Rape' in crime_type:
        serious_crime = True
    
    # denoting assault/abuse related crime
    if 'Assault' in crime_type or 'Abuse' in crime_type or "Asslt" in crime_type:
        serious_crime = True
    
    # denoting theft with weapon related crime
    if 'Robb' in crime_type:
        if 'Firearm' in crime_type or "Knife" in crime_type or 'Wpn' in crime_type:
            serious_crime = True
    
    # denoting murder related crimes
    if 'Murder' in crime_type:
        serious_crinme = True
        
    return serious_crime
```

## Displaying Map with Crime Types Popped Out

```python
arrest_table['Serious'] = arrest_table['incidentOffense'].apply(differentiator)
arrest_table
```

```python
for i, row in arrest_table.sample(300).iterrows():
    if row['Serious']:
        folium.Marker(
            [row['lat'], row['long']],
            popup=row['incidentOffense'],
            icon=folium.Icon(color='red'),
        ).add_to(map_osm)
    else:
        folium.Marker(
            [row['lat'], row['long']],
            popup=row['incidentOffense'],
            icon=folium.Icon(color='blue'),
        ).add_to(map_osm)

map_osm
```

As we can see after zooming in, it seems as though there is a greater proliferation of **non-serious** crimes (as defined above) as opposed to **serious** crimes (as defined above) when we take a random sample of 300 crimes that have occurred in this data set. However, in doing greater data exploration, it seems as though **narcotics** related crimes have a large proliferation within this data set. Given that the legalization of marijuana is imminent in the US, we should isolate marijuana related cases so as to see to what degree, per this data set, has the Baltimore Police been utilising its police force to enforce Marijuana related crimes as opposed to other crimes. 

```python
# resetting the map
map_osm = folium.Map(location=[39.29, -76.61], zoom_start=11)


# generating a new differentiator that segements crimes based on whether they are related to marijuana or not
def weed_diff(charge):
    if "Marihuana" in charge or "Marijuana" in charge:
        return True
    else:
        return False
```

```python
# making a new table related to weed-related crimes
arrest_table['chargeDescription'].fillna('', inplace=True)
arrest_table['Weed Related'] = arrest_table['chargeDescription'].apply(weed_diff)
arrest_table
```

```python
for i, row in arrest_table.sample(300).iterrows():
    if row['Serious']:
        folium.Marker(
            [row['lat'], row['long']],
            popup=row['incidentOffense'],
            icon=folium.Icon(color='red'),
        ).add_to(map_osm)
    else:
        if row['Weed Related']:
            folium.Marker(
                [row['lat'], row['long']],
                popup=row['incidentOffense'],
                icon=folium.Icon(color='green'),
            ).add_to(map_osm)
        else:
            folium.Marker(
                [row['lat'], row['long']],
                popup=row['incidentOffense'],
                icon=folium.Icon(color='blue'),
            ).add_to(map_osm)

map_osm
```

As we can see from the map above, it seems as though, from a random sample of 300 crimes, a large visible percentage of **non-serious** crimes are crimes related to **marijuana possession**. This naturally brings into question the efficacy of law enforcement, to what degree should we try to prosecute marijuana-related crimes over other **non-serious** and **seroius** crimes.
